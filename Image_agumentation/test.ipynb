```python
import pandas as pd
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
import os



```

```python
''' Function for dataframe vectorization '''

def change_rect(rect):
    x, y, w, h = rect
    return [int(x), int(y), int(x+w), int(y+h)]
    
from imgaug.augmentables.bbs import BoundingBox, BoundingBoxesOnImage
def bounrects(rects):
    return [BoundingBox(x1=rect[0], y1=rect[1], x2=rect[2], y2=rect[3]) for rect in rects]


def x1 (rect ):
    return rect[0]

def y1 (rect ):
    return rect[1]

def x2 (rect ):
    return rect[2] + rect[0]

def y2 (rect ):
    return rect[3] + rect[1]
    
```

```python
import glob
from imgaug.augmentables.bbs import BoundingBox, BoundingBoxesOnImage

img_id = 0
anno_id = 0
for img_path in glob.glob(r"D:\datasets\image_agumentation\\*\*\*\*e.png"):
    class_name = img_path.split('\\')[-4]
    date = img_path.split('\\')[-2]
    image = cv.imread(img_path)
    
    print(img_path)

    
    # print(json_f)
    df = pd.read_pickle(img_path[0:-17]+'\output.pkl')
    df['rect'] = df.straight_rect.apply(change_rect)
    df['x1'] = df.rect.apply(x1)
    df['y1'] = df.rect.apply(y1)
    df['x2'] = df.rect.apply(x2)
    df['y2'] = df.rect.apply(y2)
    



    # print(type(json_f['annotations']))

    l = 512
    b = 512
    d = 0
    for i in range(1,8):
        for j in range(1,6):

            b_end = i*l
            l_end = j*b
            b_start = (i-1)* l
            l_start = (j-1)* b

            dfing = df[(df.x1 >= b_start) & (df.x1 <= b_end) & (df.y1 >= l_start) & (df.y1 <= l_end)]
            if l_end> image.shape[0]:
                l_end = image.shape[0]
            if j*b> image.shape[1]:
                b_end = image.shape[1]

            img_t = image[b_start:b_end,l_start:l_end]
            try:
                plt.imshow(img_t)
            except:
                print('error')

            newdir_path = img_path[0:-18]+' ' + str(d) +'\img_aug'
            os.makedirs(newdir_path)

            name = newdir_path[0:-7]+str(d) +"imagea.png"
            try:
                cv.imwrite(name,img_t)
            except:
                print('error')

            dfing.to_pickle(newdir_path[0:-7]+'output.pkl')

            d+=1
        

```
